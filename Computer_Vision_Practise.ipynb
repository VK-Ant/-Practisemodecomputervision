```python
#edgedetector(canny),dialation(image edgethinkness),opposite of dialation is eroation
import cv2
import numpy as np

image = cv2.imread("ANY.tiff")
kernel = np.ones((5,5),np.uint8)

convert = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)#this is gray image covert image

blur = cv2.GaussianBlur(convert,(7,7),0) #img,kernel,sigmax

cannyedge = cv2.Canny(image,100,100) #thersholdvalue for 100,100

dialation= cv2.dilate(cannyedge,kernel,iterations=1) #imageedgethinkness image,kernel,howmanytimeruns

erotion = cv2.erode(dialation,kernel,iterations = 1)#(thiner)


cv2.imshow("edgedetectorusingcanny",cannyedge)
cv2.imshow("GrayscaleImage",convert)
cv2.imshow("blurimage",blur)
cv2.imshow("orginalimage",image)
cv2.imshow("dialationimage",dialation)
cv2.imshow("erotionImage",erotion)
cv2.waitKey(0)
```

```python
#wrap presepctive
import numpy as np
import cv2
card = cv2.imread("images.jfif")
width,height = 250,350

pts1 = np.float32([[111,219],[287,188],[154,482],[352,440]])
#this is the right,left,up and down points
pts2 = np.float32([[0,0],[width,0],[0,height],[width,height]])
matrix = cv2.getPerspectiveTransform(pts1,pts2)
imgoutput = cv2.warpPerspective(card,matrix,(width,height))


cv2.imshow("output",imgoutput)
cv2.imshow('Image',card)
cv2.waitKey(0)
```

```python
#COLOR DETECTION using trackbar

import cv2
import numpy as np



def empty():  #emptyfunction
    pass



cv2.namedWindow("TrackBars")  # create the name
cv2.resizeWindow("TrackBars",640,240)  #create the size of output trackbar window
cv2.createTrackbar("Hue Min","TrackBars",0,179,empty) #create hue max,min &sat max,min&val max min
cv2.createTrackbar("Hue Max","TrackBars",19,179,empty)
cv2.createTrackbar("sat Min","TrackBars",110,255,empty)
cv2.createTrackbar("sat Max","TrackBars",240,255,empty)
cv2.createTrackbar("val Min","TrackBars",153,255,empty)
cv2.createTrackbar("val Max","TrackBars",255,255,empty)





while True: #loop the true the all function
    path= cv2.imread("ANY.tiff") #read the input
    imghsv = cv2.cvtColor(path,cv2.COLOR_BGR2HSV) #cvt the normal to hsv
    h_min = cv2.getTrackbarPos("Hue Min","TrackBars")# position of the image to change the color for all the trackbar
    h_max = cv2.getTrackbarPos("Hue Max","TrackBars")
    s_min = cv2.getTrackbarPos("sat Min","TrackBars")
    s_max = cv2.getTrackbarPos("sat Max","TrackBars")
    v_min = cv2.getTrackbarPos("val Min","TrackBars")
    v_max = cv2.getTrackbarPos("val Max","TrackBars")

    print(h_min,h_max,s_min,s_max,v_min,v_max)#print the value of color
    lower = np.array([h_min,s_min,v_min])
    upper = np.array([h_max,s_max,v_max])
    mask = cv2.inRange(imghsv,lower,upper)#the mask is 0 and 1(black,white), then trackbar change how to position the color
    result = cv2.bitwise_and(path,path,mask=mask)

    cv2.imshow("originalimage",path)
    cv2.imshow("HSVimage",imghsv)
    cv2.imshow("maskimage",mask)
    
    cv2.imshow("Result",result)
    
    
    cv2.waitKey(1)
```

```python
#FaceDetection
#viola and jones(one of the earilest method for real time object detection)

#Positive's faces + negative's faces ------> Train ------> XML File
#opencv cascades is provide easily detect the face,eye
#1 haarcascade_eye etc........

import cv2

faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")  #its is the haarcascade frontal face is predefined XML file
image = cv2.imread("4.1.03.tiff")
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)


faces = faceCascade.detectMultiScale(gray,1.1,4)
#This is the face detect the multiscale face detection for gray and scalingfactor,imagefactor

for(x,y,w,h) in faces: 
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,255,255),5)     #This is the x,y,h,w --> mainly focused on bounding box of image
    

cv2.imshow("Image",image)
cv2.waitKey(0)



```
